```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')
```

```python
train_data.iloc[:,1].replace("[^a-zA-Z]"," ",regex=True, inplace=True)
train_data["Review"]=train_data["Review"].str.lower()
```

```python
test_data.iloc[:,1].replace("[^a-zA-Z]"," ",regex=True, inplace=True)
test_data["Review"]=test_data["Review"].str.lower()
```

```python
COMMENT = 'Review'
train_data[COMMENT].fillna("unknown", inplace=True)
test_data[COMMENT].fillna("unknown", inplace=True)
```

```python
headlines = []
for row in range(0,len(train_data.index)):
    headlines.append(train_data.iloc[row,1])
```

```python
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(1,1))
traindataset=countvector.fit_transform(headlines)
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train_data['Polarity'])
```

```python
headlines_test = []
for row in range(0,len(test_data.index)):
    headlines_test.append(test_data.iloc[row,1])
```

```python
headlines_test[0]
```

```python
test_dataset = countvector.transform(headlines_test)
```

```python
predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
test_data['Polarity']=pd.DataFrame(predictions)
```
